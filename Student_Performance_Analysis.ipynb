```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: nicer plots
sns.set(style="whitegrid")
```

```python
# Load dataset with semicolon separator
df = pd.read_csv(r"C:\Users\Admin\Downloads\student\student-mat.csv", sep=';')

# Check first 5 rows
df.head()
```

```python
df.isnull().sum()
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
## Data Cleaning
#- Checked missing values
#- Removed duplicate records
#- Verified data types
```

```python
#  Average final grade (G3)
average_g3 = df['G3'].mean()
print("Average Final Grade (G3):", average_g3)
```

```python
#  Students scored above 15
above_15 = df[df['G3'] > 15].shape[0]
print("Number of students scoring above 15:", above_15)
```

```python
# Correlation: study time vs final grade
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and G3:", correlation)
```

```python
# Average performance by gender
gender_avg = df.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender:\n", gender_avg)
```

```python
#Histogram of Grades
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Scatter Plot: Study Time vs Grades
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'], alpha=0.7, color='green')
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (hours/week)")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
#Bar Chart: Gender vs Average Score
gender_avg.plot(kind='bar', color=['blue', 'pink'])
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

```python
## Conclusion
#- The average student grade is around 11-12.  
#- Students who study more tend to score higher (positive correlation).
#- Female students perform slightly better on average than male students.  
#- Visualizations help understand the trends in student performance clearly.
```
